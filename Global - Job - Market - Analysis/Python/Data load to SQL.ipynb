```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "mysql+mysqlconnector://root:shreya25@localhost:3306/job_market"
)
```

```python
df = pd.read_csv("main_job_postings_global_skills.csv")

df.to_sql(
    "job_postings",
    con=engine,
    if_exists="replace",
    index=False,
    chunksize=5000
)

print("✅ Table created + Data inserted successfully")
```
